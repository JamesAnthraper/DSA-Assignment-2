## 3. Plotting Bar graph of male vs female buyers

```python
fig=data['Buyer Gender'].value_counts().to_frame().reset_index()
plt.bar(fig['index'],fig['Buyer Gender'], color=['#FF69B4','#00BFFF'])
plt.title('Buyers Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
```

#### Countplot

```python
sns.countplot(x='Buyer Gender', data=data)
```
